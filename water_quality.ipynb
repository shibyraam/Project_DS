```python
# fill the nan values, ( filled in (median value) center number of values filled on nan values)

# df['ph'].fillna(df['ph'].median(), inplace=True)
# df['Sulfate'].fillna(df['Sulfate'].median(), inplace=True)
# df['Trihalomethanes'].fillna(df['Trihalomethanes'].median(), inplace=True)

# Imputing 'ph' value 

phMean_0 = df[df['Potability'] == 0]['ph'].median(skipna=True)
df.loc[(df['Potability'] == 0) & (df['ph'].isna()), 'ph'] = phMean_0
phMean_1 = df[df['Potability'] == 1]['ph'].median(skipna=True)
df.loc[(df['Potability'] == 1) & (df['ph'].isna()), 'ph'] = phMean_1

# Imputing 'Sulfate' value 

SulfateMean_0 = df[df['Potability'] == 0]['Sulfate'].median(skipna=True)
df.loc[(df['Potability'] == 0) & (df['Sulfate'].isna()), 'Sulfate'] = SulfateMean_0
SulfateMean_1 = df[df['Potability'] == 1]['Sulfate'].median(skipna=True)
df.loc[(df['Potability'] == 1) & (df['Sulfate'].isna()), 'Sulfate'] = SulfateMean_1

# Imputing 'Trihalomethanes' value 

TrihalomethanesMean_0 = df[df['Potability'] == 0]['Trihalomethanes'].median(skipna=True)
df.loc[(df['Potability'] == 0) & (df['Trihalomethanes'].isna()), 'Trihalomethanes'] = TrihalomethanesMean_0
TrihalomethanesMean_1 = df[df['Potability'] == 1]['Trihalomethanes'].median(skipna=True)
df.loc[(df['Potability'] == 1) & (df['Trihalomethanes'].isna()), 'Trihalomethanes'] = TrihalomethanesMean_1
```

```python

q1 = df.quantile(0.25)
q1
```

```python
IQR = q3-q1
IQR
```

```python
# Lower Range
lower_range = q1 - 1.5*IQR
upper_range = q3 + 1.5*IQR
```

```python
lower_range
```

```python
upper_range
```

```python
df1 = df[(df>lower_range)&(df<upper_range)]
df1
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
from sklearn.impute import KNNImputer
imputer= KNNImputer()
df2=imputer.fit_transform(df1)
df2
```

```python
df2_column = imputer.get_feature_names_out()
df2_column
```

```python
df2 = pd.DataFrame(df2)
df2.columns = df2_column
df2
```

```python
plt.figure(figsize=(8,7))
sns.heatmap(df2.corr(),annot=True).set_title('water_potability')
```

```python
# processed csv store in this path
df2.to_csv('processed_csv/processed_file.csv')
```

```python
# Lable selection
data=df2.copy()
X = data.drop('Potability',axis=1).copy()
y = data['Potability']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,shuffle=True, random_state=101)
```

```python
X_train.to_csv('train/xtrain.csv')
y_train.to_csv('train/ytrain.csv')
```

```python
X_test.to_csv("test/xtest.csv")
y_test.to_csv("test/ytest.csv")
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(X_train.corr(),annot=True)
```

```python
bestfeature = SelectKBest(score_func=chi2, k=9)
fit =  bestfeature.fit(X_train,y_train)
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
featurescores = pd.concat([dfcolumns,dfscores],axis=1)
featurescores.columns = ['specs','score']
print(featurescores.nlargest(9, 'score'))
```

```python
feature = featurescores.nlargest(9, 'score')
feature
```

```python
feature_list = feature['specs'].to_list()
```

```python
feature_list
```

```python
jb.dump(feature_list, 'train/model/featureselected.pkl')
```

```python
feature_selected = X_train[feature_list]
```

```python
feature_selected
```

```python
feature_selected.to_csv('train/processed/featureselected_X_train.csv',index=False)
```
